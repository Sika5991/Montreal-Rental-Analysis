# Part 1: Scrap Kijiji Rental Data

This program is designed to scrape information from Kijiji ad URLs, extracting various details such as ad ID, utility status, price, size, and more. The objective of the code is to accumulate a sufficiently large sample size of rental data to facilitate an in-depth analysis.

source: https://www12.statcan.gc.ca/census-recensement/2021/dp-pd/prof/details/page.cfm?Lang=E&SearchText=montreal&DGUIDlist=2021S0503462&GENDERlist=1,2,3&STATISTIClist=1,4&HEADERlist=0

```python
#import libraries
from bs4 import BeautifulSoup as bs
import numpy as np
import requests
import re
import requests
from time import sleep
from random import randint
import pandas as pd
import os
import csv
import numpy
```

```python
# Base url
base_url = 'https://www.kijiji.ca'
language_url = '?siteLocale=en_CA'

# Iterate over pages
pages = np.arange(70)

# Store ad url's in a list
ad_list = []

# Define list
kijiji_list = []

# Define dictionnairy
kijiji_dic = {}
```

```python

kijiji_list.clear()
print(kijiji_list, kijiji_dic)
```

```python
#Scrap kijiji for ad URLS

for page in pages:
    url = 'https://www.kijiji.ca/b-apartments-condos/grand-montreal/page-' + str(page) + '/c37l80002?siteLocale=en_CA&sort=dateDesc'
    r_url = requests.get(url)
    soup = bs(r_url.content)
    sleep(randint(2, 10))
    
    ad_links = soup.find_all('h3', class_='sc-c54bbc09-0 KhHgs sc-379c64e5-13 knNVTD')

    for ad_link in ad_links:
        link_tag = ad_link.find('a', class_='sc-379c64e5-0 dlxsIh')
        if link_tag:
            ad_url = base_url + link_tag['href'] + language_url
            ad_list.append(ad_url)
            print(f"Captured URL: {ad_url}")
```

```python
# Initialize a set to keep track of ad IDs
visited_ad_ids = set()

# Loop through ad URLs
for ad_url in ad_list:
    # Scrap ad ID
    r_specific = requests.get(ad_url)
    soup = bs(r_specific.content, 'html.parser')
    try:
        content_key_ad_element = soup.find('li', class_='currentCrumb-2810804557')
        if content_key_ad_element:
            content_key_ad = content_key_ad_element.get_text().split()[0]
            ad_id = content_key_ad_element.get_text().split()[-1]
            kijiji_dic = {'ad_id': ad_id}
            # Check if ad ID is already visited, if so, skip scraping
            if ad_id in visited_ad_ids:
                continue
            visited_ad_ids.add(ad_id)  # Add the ad ID to the set of visited ad IDs
        else:
            kijiji_dic = {'ad_id': 'NA'}
    except AttributeError:
        kijiji_dic = {'ad_id': 'NA'}
    
    # Scrap price and utility status
    for row in soup.find_all('div', class_='priceWrapper-3915768379'):
        if row is not None:
            price = row.span.get_text(strip=True).replace('\xa0', '')
            kijiji_dic['price'] = price if price else 'NA'
            
            utility_flag_element = row.find('span', class_='utilities-3805585157')
            if utility_flag_element:
                utility_flag = utility_flag_element.get_text()
            else:
                utility_flag = 'NA'
            kijiji_dic['utility_flag'] = utility_flag if utility_flag else 'NA'

    # Scrap address
    location_row = soup.find('div', class_='locationRow-3895896735')
    address = location_row.find('span', class_='address-2094065249').get_text() if location_row else 'NA'
    kijiji_dic['address'] = address if address else 'NA'


    # Scrap unit details
    unit_test = soup.find('div', class_='titleAttributes-183069789')
    if unit_test:
        for index, row in enumerate(unit_test):
            text = row.get_text()
            if index == 0:
                if not text.strip():
                    text = "NA"  # Replace empty text with "NA"
                kijiji_dic['unit_type'] = text
            else:
                content_split = text.split(':')
                if len(content_split) == 2:
                    content_key_unit = content_split[0].strip()
                    content_value_unit = content_split[1].strip()
                else:
                    content_key_unit = text.strip()
                    content_value_unit = ""

                # Check if content_value_unit is empty and replace with "NA"
                if not content_value_unit:
                    content_value_unit = "NA"
                kijiji_dic[content_key_unit] = content_value_unit
    
    attribute_group = soup.find_all('li', 'attributeGroupContainer-1655609067')

    # Scrap utility items
    content_keys = []
    utilities = []
    for content in attribute_group:
        content_keys_element = content.find('h4', class_='attributeGroupTitle-889029213')
        if content_keys_element:
            content_key = content_keys_element.get_text()
            content_keys.append(content_key)
            content_values = content.find_all('li', class_='groupItem-1182798569 available-1766233427')
            values_list = [value.get_text() for value in content_values]
            kijiji_dic[content_key] = values_list

    # Scrap unit attributes
    two_line_attributes = soup.find_all('li', class_='twoLinesAttribute-633292638')
    for row in two_line_attributes:
        content_key = row.find('dt', class_='twoLinesLabel-2332083105').get_text() if row.find('dt', class_='twoLinesLabel-2332083105') else 'NA'
        content_value = row.find('dd', class_='twoLinesValue-2653438411').get_text() if row.find('dd', class_='twoLinesValue-2653438411') else 'NA'
        kijiji_dic[content_key] = content_value if content_value else 'NA'

    # Add ad_url
    kijiji_dic['ad_url'] = ad_url
    print(kijiji_dic)

    # Add to list
    kijiji_list.append(kijiji_dic)
```
