Import necessary libraries

```python
from bs4 import BeautifulSoup as bs
import requests
import re
```

Load Montreal Kijiji webpage

```python
r = requests.get("https://www.kijiji.ca/v-appartement-condo/longueuil-rive-sud/logements-a-louer-a-longueuil/1554985109?siteLocale=en_CA")

#Convert into soup
soup = bs(r.content)

#Print html
print(soup.prettify())
```

```python
#Get a specific info on a listing

r_specific = requests.get('https://www.kijiji.ca/v-appartement-condo/longueuil-rive-sud/logements-a-louer-a-longueuil/1554985109?siteLocale=en_CA')

#convert into soup
soup = bs(r_specific.content)

print(soup.prettify())
```

```python
results = []
data_dic = {}

# Scrape the ad ID
ad_id = soup.find('a', class_= 'adId-1017171818').get_text(strip= True).replace('\xa0', '')
data_dic['ad_id'] = ad_id
#print(results)


#scrap price and  utility status
for row in soup.find_all('div', class_= 'priceWrapper-3915768379'):
    data_dic['price'] = row.span.get_text(strip=True).replace('\xa0', '')
    data_dic['utility_flag'] = row.find('span', class_='utilities-3805585157').get_text()
    
print(data_dic)
```

```python
#scrape location
location_row = soup.find('div', class_= 'locationRow-3895896735')
location = location_row.find('span', class_= 'address-2094065249').get_text()
print(location)
```

```python
#scrape unit type, number of bedrooms and number of bathrooms
unit_details_rows = soup.find('div', class_= 'titleAttributes-183069789')

unit_details = unit_details_rows.find_all_next('span', class_= 'noLabelValue-774086477')

for index, row in enumerate(unit_details):
    text_content = row.get_text(strip=True)
    if index == 0:
            # Extract unit type
            data_dic['unit_type'] = text_content
    elif index == 1:
            # Extract number of bedrooms
            data_dic['number_bedrooms'] = re.search(r'\d+$',text_content).group()
    elif index == 2:
            # Extract number of bathrooms
            data_dic['number_bathrooms'] = re.search(r'\d+$', text_content).group()
    
print(data_dic)
```
